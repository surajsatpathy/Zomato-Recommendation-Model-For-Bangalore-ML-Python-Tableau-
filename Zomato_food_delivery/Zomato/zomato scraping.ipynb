```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver import Firefox
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import service
```

```python
url='https://www.zomato.com/bangalore/restaurants'
```

```python
browser=webdriver.Chrome()
browser.get(url)
```

```python
html=browser.page_source
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
rest=soup.find_all('div',{'class':'jumbo-tracker'})
rest[0].text
```

```python
restaurant_name=[]
ratings=[]
price_for_one=[]
cuisins=[]
link=[]

for i in soup.find_all('h4',{'class':'sc-1hp8d8a-0 sc-vBKru ktSVmG'}):
    restaurant_name.append(i.text)
for j in soup.find_all('div',{'class':'sc-1q7bklc-1 cILgox'}):
    ratings.append(j.text)
for k in soup.find_all('p',{'class':'sc-1hez2tp-0 sc-kIXKos litzDQ'}):
    price_for_one.append(k.text)
for m in soup.find_all('p',{'class':'sc-1hez2tp-0 sc-kIXKos dCJcaq'}):
    cuisins.append(m.text)
for l in soup.find_all('a', href = True):
    if 'order' in l['href']:
        if l['href'] not in link:
            
            link.append(l['href'])
```

```python
delivery_review_no=[]
locations=[]
count = 0
for i in link:
    count+=1
    print(count,end=" ")
    r=browser.get('https://www.zomato.com/'+i)
    sou = BeautifulSoup(browser.page_source, 'lxml')
    try:
        j = sou.find_all("a",{"class" : "sc-fZwumE gIYZim"})
        locations.append(j[0].text)
    except:
        locations.append('NA')
        
    try:
        j = sou.find_all("div",{"class" : "sc-1q7bklc-8 kEgyiI"})
        delivery_review_no.append(j[0].text)
    except:
        delivery_review_no.append('NA')
```

```python
a={'restaurant_name':restaurant_name,'ratings':ratings,'price_for_once':price_for_one,'cuisins':cuisins,'links':link,
    'locations': locations,'delivery_review_number':delivery_review_no
    
}
```

```python
a2=pd.DataFrame(a)
```
